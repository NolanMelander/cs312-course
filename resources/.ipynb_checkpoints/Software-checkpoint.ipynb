## 1 - Python 3

All of the coding in CS312 will be implimented using Python version 3.  Python will be downloaded using the python package manager conda. (See the next section).  Here is the Python 3 [tutorial](https://docs.python.org/3/tutorial/) and [interactive Python tutorial](http://www.learnpython.org/).


## 2 - AnaConda

AnaConda is a python package manager.  Python and all of it's libraries can be difficult to manage (ie., updating a library can cause other libraries to fail).  We will be using the full version of AnaConda for this course.

### 2.1 - Installing Conda

- Download anaconda at [anaconda](https://anaconda.org).  We need the 64 bit version that has the GUI.
- While installing, make sure that the option to register Anaconda as my default Python 3.6. 

![](../images/installconda.png)

### 2.2 - Install Components Required for the Course

On your computer, search for "AnaConda Prompt".  This is a terminal window where we will use AnaConda to install libraries and run software.  (Mac users should just open a terminal window to type the following commands).  Type the following and execute one at a time.

```
    conda install -c conda-forge opencvÂ 

    conda install matplotlib

    conda install jupyter
```


## 3 - Running Jupyter Notebooks

All of the course matterial and assignments are written in [Jupyter Notebook](http://jupyter.org/).  Jupyter notebooks can be viewed directly in GitHub (such as this page).  For the course team activities and assignments, you will be required to be able to modfiy notebooks.  The above steps for Conda installed all of the software that you require.

To start Jupyter Notebook:

1. Open Anaconda Navigator.
1. From the main screen, click on the button to launch a notebook.
1. A page will be displayed in your computer's Internet browser.  This is the main page of Jupyter.
1. Navigate to where the course notebooks are located on your computer.
1. When you finished running notebooks, exit the AnaConda Navigator.


## 4 - Visual Studio Code

During the course, we might be using the text editor [Visual Studio Code](https://code.visualstudio.com/).  It was selected because it's free and can run on Windows, Mac and Linux.  We are free to use any editor that you want.

### 4.1 - Connecting AnaConda and VSCode

In order to connect VSCode to AnaConda you must add the path of AnaConda in the settings file of VSCode.  Go to Workplace Settings by pressing Cmd , or navigating from the top menu, then add this to settings.json.  Below is a MAC OS example.


**MAC OS**
```
{
    "python.pythonPath": "~/anaconda/envs/${env.CONDA_DEFAULT_ENV}/bin/python"
}
```

**Windows (where NNNN is your user name)**
```
{
    "python.pythonPath": "c:\\users\\NNNN\\anaconda3"
}

```
